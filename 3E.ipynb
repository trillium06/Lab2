```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
F0 = 5.00e2             # N
k = 6.40                # N/m
f0 = 60.0               # Hz
omega0 = 2 * np.pi * f0 # rad/s
m = k / omega0**2       # kg

# Damping widths (FWHM)
df1 = 5.0   # Hz
df2 = 2.5   # Hz (narrower peak)
gamma1 = 2 * np.pi * df1
gamma2 = 2 * np.pi * df2

# Frequency range
f = np.linspace(40, 80, 1000)
omega = 2 * np.pi * f

# Average power for both cases
P1 = 0.5 * (F0**2 / m) * (gamma1 * omega**2) / ((omega0**2 - omega**2)**2 + (gamma1 * omega)**2)
P2 = 0.5 * (F0**2 / m) * (gamma2 * omega**2) / ((omega0**2 - omega**2)**2 + (gamma2 * omega)**2)

# Maximum power at resonance
Pmax1 = 0.5 * (F0**2 / (m * gamma1 * omega0))
Pmax2 = 0.5 * (F0**2 / (m * gamma2 * omega0))

# Plot
plt.figure(figsize=(10, 6))
plt.plot(f, P1, label=f"FWHM = 5.0 Hz, P = {Pmax1:.2e} W", color='royalblue')
plt.plot(f, P2, label=f"FWHM = 2.5 Hz, P = {Pmax2:.2e} W", color='crimson')
plt.axvline(f0, color='orange', linestyle='--', label="Resonant frequency (60 Hz)")
plt.title("Resonance Peaks for Different Damping Widths")
plt.xlabel("Driving Frequency f (Hz)")
plt.ylabel("Average Power P (W)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print(f" = {gamma1:.3f} rad/s  Pmax = {Pmax1:.3e} W")
print(f" = {gamma2:.3f} rad/s  Pmax = {Pmax2:.3e} W")
```
