```python
import numpy as np
import matplotlib.pyplot as plt

# Given constants
omega_0 = 4.00       # natural frequency
omega_d = 5.00       # driving frequency
A0 = 5.00            # amplitude of homogeneous solution
kappa = 10.0         # driving amplitude

# Time array (0 to ~25 seconds to show beating behavior)
t = np.linspace(0, 25, 5000)

# Solution for x(t)
x = A0 * np.cos(omega_0 * t) + (kappa / (omega_0**2 - omega_d**2)) * np.cos(omega_d * t)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(t, x, color='royalblue')
plt.title("Driven, Undamped Oscillator (  _d)")
plt.xlabel("Time t (s)")
plt.ylabel("Displacement x(t)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
